```python
from transformers import AutoTokenizer, AutoModelForCausalLM

# AutoModelForCausalLM transformer for next sentence predictions
```

```python
# Hugging Face Model
# https://huggingface.co/google/gemma-3-1b-it

model_name = "google/gemma-3-1b-it"
```

```python
# Make a tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Pulls tokenizer files from hugging face
```

```python
input_prompt = [
    "The capital of India is"
]

# Goal: want AutoModelForCausalLM gemma-3 to compelte this sentence (initial string)
```

```python
# Tokenise the input
tokenized = tokenizer(input_prompt, return_tensors="pt")

tokenized["input_ids"]
```

```python
# Load the model
import torch

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.bfloat16
)
```

```python
gen_result = model.generate(tokenized["input_ids"], max_new_tokens=25)

gen_result # predicted the next set of tokens
```

```python
output = tokenizer.batch_decode(gen_result)
output
```
